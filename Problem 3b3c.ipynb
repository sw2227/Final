```
import numpy as np
import matplotlib.pyplot as plt

# Final - Problem 3 - (b)
# Originally from Prelim 1
n_old = np.matrix([
                 [0,19,18,20],    
                 [0.0005,21,17,26],
                 [0.005,41,37,44],
                 [0.012,67,65,69],
                 [0.053,86,84,88],
                 [0.216,93,91,95],
                 [1,93,92,94]
                 ])

n_new = n_old
IPTG = n_new[:,0]
#Numbers from Prelim 1 Solution
W1 = 0.25;
n = 1.85;
K = 9e-2;
W2 = 98.75;

fI = (np.power(IPTG, n))/(np.power(K, n) + np.power(IPTG, n))
ui=(W1+W2*fI)/(1+W1+W2*fI)
pi = 0.575*1.4204*ui
plt.figure(figsize=(7, 7))
plt.plot(ui, pi,'r--');
plt.xlabel(r'$\bar{ui}$', fontsize = 15);
plt.ylabel('pi* (nmol/gDW)', fontsize = 15);
```

```
#Problem 3 -(c)
fI = (np.power(IPTG, n))/(np.power(K, n) + np.power(IPTG, n))
ui=(W1+W2*fI)/(1+W1+W2*fI)
pi = 0.575*1.4204*ui
pi_new1 = 5*0.575*1.4204*ui #Kp>1, choose Kp = 5
pi_new2 = 1.1*0.575*1.4204*ui #Kp>1, choose Kp = 1.1

plt.figure(figsize=(7, 7))
plt.plot(ui, pi_new1,'c--', label = 'Kp = 5');
plt.plot(ui, pi_new2,'b--', label = 'Kp = 1.1');
plt.plot(ui, pi,'r--', label = 'Kp = 1');
plt.xlabel(r'$\bar{ui}$', fontsize = 15);
plt.ylabel('pi* (nmol/gDW)', fontsize = 15);
plt.legend(loc = 'upper left');
```

As one can observe from the figure above, the curve moved up when Kp increased. This makes sense because as the polysome amplification constant grows larger, there will be more ribosomes reading a message at any time, thus resulting in more translation activity and as a result, more intracellular proteins.(Assuming there are sufficient mRNAs to translate)
