```
import numpy as n
import matplotlib.pyplot as p
#Probelm 1 - (b)
m = 50
V1 = n.linspace(0, 1.1, m)
V2 = n.linspace(0, 1.1, m)
M1, M2 = n.meshgrid(V1, V2)
D1t = n.zeros(M1.shape)
D2t = n.zeros(M2.shape)

def calc(dd):
    D1, D2 = dd
    y = 1/(1 + 10*n.power(n.power(D2, 2)/(0.1 + n.power(D2, 2)), 2)) - D1
    z = 1/(1 + 10*n.power(n.power(D1, 2)/(0.1 + n.power(D1, 2)), 2)) - D2
    return [y, z]

for a in range(m):
    for b in range(m):
        D1m = M1[a, b]
        D2m = M2[a, b]
        ans = calc([D1m, D2m])
        D1t[a, b] = ans[0]
        D2t[a, b] = ans[1]

p.figure(figsize = (10, 10))
p.quiver(M1, M2, D1t, D2t, color = 'black');
p.xlabel('D1');
p.ylabel('D2');
```

```
# Adding nullclines
r = n.linspace(0, 1.1, 200)
nullcline = 1/(1 + 10*n.power(n.power(r, 2)/(0.1 + n.power(r, 2)), 2))
     
p.figure(figsize = (20, 20))
p.quiver(M1, M2, D1t, D2t, color = 'black');
p.plot(r, nullcline, color = 'c');
p.plot(nullcline, r, color = 'r');
p.xlabel('D1');
p.ylabel('D2');
```

One can observe three steady states from the three intersections of the nullclines. By observing the direction of the arrows, one can see that there are two stable steady states and one unstable steady state. In a long-time limit, the system will settle into these two stable steady states and the cell with higher Delta concentration will assume the primary fate, while the other with assume the secondary fate. This result is similar to the case discussed in lecture.
