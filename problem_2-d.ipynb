```
import numpy as np
import matplotlib.pyplot as plt
# Problem 2 - (d)
ke = 0.0001;
ke2 = 0.005;
kf = 5.14e-21; 
kr = 2.5e-2; 
kdeg = 8e-4;
Vs = 18;
q = 1000;
nc = 3e8;
r = 100;
DL = 1e-10;
Kss = ke2*kf/(ke*(kr + ke2))
z = np.linspace(0.00001, 1.0, 10000)
km = (np.power(DL,2/3))*(np.power(r,1/3))*(np.power(z,-1/3))
RTot = Kss*q*nc*Vs*(1/ke2 + 1/kdeg)/((Kss*Vs*nc)+km)

plt.plot(z, RTot)
plt.title('Mitotic activity profile with z')
plt.xlabel('z');
plt.ylabel('Mitotic Activity');
```

```
RTot_norm = RTot/(np.amax(RTot))
plt.plot(z, RTot_norm)
plt.title('Normalized mitotic activity profile with z')
plt.xlabel('z');
plt.ylabel('Normalized mitotic Activity');
```
